# Example 3.7: Spectral density of transformed processes

In this example, we will look at the spectral density of transformed processes through matrix transformations.

### Problem

 Let $\mathbf{S}_\mathbf{x}(\omega)$ be the $2\times 2$ cross-spectral density matrix of $\mathbf{x}(t)$:

$$
\mathbf{x}(t)=
\left[
\begin{array}{c}
x_1(t) \\
x_2(t) \\
\end{array}
\right]
,\
\quad
\mathbf{S}_\mathbf{x}(\omega)=
\left[
\begin{array}{c}
S_{X_1 X_1}(\omega) & S_{X_1 X_2}(\omega) \\
S_{X_2 X_1}(\omega) & S_{X_2 X_2}(\omega)\\
\end{array}
\right]
$$

where $x_1(t)$ and $x_1(t)$ are two random processes. Let $\mathbf{\tilde{x}}(t)$ be a transformation of $\mathbf{x}(t)$:

$$
\mathbf{\tilde{x}}(t)=
\left[
\begin{array}{c}
\tilde{x}_1(t) \\
\tilde{x}_2(t) \\
\end{array}
\right]
=
\mathbf{T}
\mathbf{x}(t)
$$

where $\mathbf{T}$ is a $2\times2$ transformation matrix. We will show that the cross-spectral density matrix for $\mathbf{\tilde{x}}(t)$ is given by:


$$
\underbrace{\mathbf{S}_\mathbf{\tilde{x}}(\omega)}_{2\times2} \:\: = \:\:
\underbrace{\mathbf{T}}_{2\times2} \:\:
\underbrace{\mathbf{S}_{\mathbf{x}}(\omega)}_{2\times2} \:\:
\underbrace{\mathbf{T}^\text{T}}_{2\times2}
$$




### Solution

The cross-spectral density matrix is defined as:

$$
\mathbf{S}_{\tilde{\mathbf{x}}}(\omega) = \dfrac{1}{2\pi} \int_{-\infty}^{\infty} \mathbf{R}_{\tilde{\mathbf{x}}}(\tau) e^{-i \omega \tau } \:\text{d} \tau
$$


where the cross-correlation matrix is:

$$
\mathbf{R}_{\tilde{\mathbf{x}}}(\tau) = \mathbb{E}[\tilde{\mathbf{x}}(t) \tilde{\mathbf{x}}^\text{T}(t + \tau)]
$$

We can substitute $\tilde{\mathbf{x}}(t) = \mathbf{T} \mathbf{x}(t)$:

$$
\mathbf{R}_{\tilde{\mathbf{x}}}(\tau) = \mathbb{E}[\mathbf{T} \mathbf{x}(t) (\mathbf{T} \mathbf{x}(t + \tau))^\text{T}]
=\mathbf{T} \mathbb{E}[\mathbf{x}(t) \mathbf{x}^\text{T}(t + \tau)] \mathbf{T}^\text{T}
$$

Therefore, the cross-correlation matrix transforms as follows:

$$
\mathbf{R}_{\tilde{\mathbf{x}}}(\tau) = \mathbf{T} \mathbf{R}_{\mathbf{x}}(\tau) \mathbf{T}^\text{T}
$$

Next, we consider the spectral density. Note that $\mathbf{T}$ can be moved outside the integral:

$$
\mathbf{S}_{\tilde{\mathbf{x}}}(\omega) = \dfrac{1}{2\pi} \int_{-\infty}^{\infty} \mathbf{R}_{\tilde{\mathbf{x}}}(\tau) e^{-i \omega \tau}  \:\text{d} \tau
= \dfrac{1}{2\pi} \int_{-\infty}^{\infty} \mathbf{T} \mathbf{R}_{\mathbf{x}}(\tau) \mathbf{T}^\text{T} e^{-i \omega \tau} \:\text{d} \tau
= \mathbf{T} \left(\dfrac{1}{2\pi}  \int_{-\infty}^{\infty} \mathbf{R}_{\mathbf{x}}(\tau) e^{-i \omega \tau} \:\text{d} \tau \right) \mathbf{T}^\text{T}
=  \mathbf{T} \mathbf{S}_{\mathbf{x}}(\omega) \mathbf{T}^\text{T}
$$

Such transformation formulas are useful, for instance for a modal transformation.


### Validity for other transformations

The result is also valid when $\mathbf{T}$ is not a square matrix. Consider for instance the process $c(t)$:

$$
c(t)=
a \cdot x_1(t) + b \cdot x_2(t) =
\left[
\begin{array}{c}
a & b
\end{array}
\right]
\left[
\begin{array}{c}
x_1(t) \\
x_2(t) \\
\end{array}
\right] =
\mathbf{T}
\mathbf{x}(t)
$$

in which case $\mathbf{T}$ is a $1\times 2$ matrix. The transformation relation 

$$
\underbrace{S_{c}(\omega)}_{1\times1} \:\: = \:\:
\underbrace{\mathbf{T}}_{1\times2} \:\:
\underbrace{\mathbf{S}_{\mathbf{x}}(\omega)}_{2\times2} \:\:
\underbrace{\mathbf{T}^\text{T}}_{2\times1}
$$

still holds.
