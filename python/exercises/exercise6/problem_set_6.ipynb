### Problem 1: Stochastic response in a single degree of freedom system


Consider the SDOF system:

$$ m \ddot{y} + c \dot{y} + k y = x(t)  $$

Assume $m=100$ kg, $c=20$ N/(m/s), and $k=1000$ N/m. 

Let $x(t)$ be a stochastic load with a rectangular spectrum, which can be expressed as a (one-sided) spectral density:

$$
S_X(\omega) 
= \left\{ \begin{array} a 20 & \quad 
0 < \omega \leq 50
\\ 0 & \quad \text{elsewhere} \end{array} \right.
$$




**a)** Calculate the spectral density of the displacement response and plot it. Comment briefly on the result.

**b)** Simulate a time series of the load with length 10 minutes. Determine a suitable time step $\Delta t$. Find the displacement response by Newmark numerical integration, and calculate the standard deviation of the response from the time series.


**c)** Using the result from question a), find the standard deviation of the displacement. Compare it to the standard deviation found in question b), and comment on the result.
