### Problem 1: Frequency response method

Consider a single degree of freedom system defined by the equation of motion: 

$$ m \ddot{y} + c \dot{y} + k y = x(t)  $$

Assume that the mass is 100 kg, that the natural frequency is 1 Hz, and the damping ratio is 5%.

**a)** The stochastic load $x(t)$ has the following one-sided spectrum:

$$
S_x(\omega) 
= \left\{ \begin{array} a 10 & \quad 
0 < \omega \leq 50
\\ 0 & \quad \text{elsewhere} \end{array} \right.
$$

What do we call this type of spectrum? Use the spectrum to simulate a time series with a length of 10 minutes and a time step of 0.05 s.

**b)** Calculate the dynamic response $y(t)$  to the generated load history by the frequency response method using the FFT and IFFT functions. Calculate also the dynamic response using Newmark numerical integration, and compare it.




### Problem 2: Stochastic response in a single degree of freedom system


Consider the SDOF system:

$$ m \ddot{y} + c \dot{y} + k y = x(t)  $$

Assume $m=100$ kg, $c=20$ N/(m/s), and $k=1000$ N/m. 

Let $x(t)$ be a stochastic load with a rectangular spectrum, which can be expressed as a (one-sided) spectral density:

$$
S_X(\omega) 
= \left\{ \begin{array} a 20 & \quad 
0 < \omega \leq 50
\\ 0 & \quad \text{elsewhere} \end{array} \right.
$$

**a)** Calculate the spectral density of the displacement response and plot it. Comment briefly on the result.

**b)** Simulate a time series of the load with length 10 minutes. Determine a suitable time step $\Delta t$. Find the displacement response by Newmark numerical integration, and calculate the standard deviation of the response from the time series.

**c)** Using the result from question a), find the standard deviation of the displacement. Compare it to the standard deviation from the response $y(t)$ found in question b), and comment on the result.

The stochastic loads can be described by a diagonal cross-spectral density matrix:

$$ \mathbf{S}_\mathbf{X} = \left[ \begin{array}{c c}
S_{X_1}(\omega) & 0 & 0 \\
0 &S_{X_2}(\omega) & 0 \\
0 & 0 & S_{X_3}(\omega)

$$
S_X(\omega)

$$

$$ \mathbf{S}_\mathbf{\tilde{X}}(\omega)  = \mathbf{\Phi}^\text{T} \mathbf{S}_\mathbf{X}(\omega)  \mathbf{\Phi}$$

**d)** Calculate the $3 \times 3$ cross-spectral density matrix of the modal coordinate responses. Plot the result and comment on its characteristics.

### Problem 4: Theory

*i)* the auto correlation of the  velocity response, $R_{\dot{y}}(\tau)$, is given by $-\dfrac{\text{d}^2 R_y(\tau)}{\text{d} \tau^2}$.

*ii)* the spectral density of the velocity response, $S_{\dot{y}}(\omega)$, is given by $\omega^2 S_y(\omega)$.
