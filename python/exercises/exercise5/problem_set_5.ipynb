### Problem 1: Simulation of stochastic time series

In this task you will simulate time series of a stochastic process. Consider the following (one-sided) spectral density, which is used to model atmospheric turbulence in wind-resistant design of bridges: 

$$
S(\omega) 
= \left\{ \begin{array} a 0 & \quad 
\omega=0\\
\dfrac{(IU)^2 A \cdot \omega_z}{\omega \left(1 + 1.5 A \cdot \omega_z \right)^{5/3}}
 & 0<\omega<\infty \end{array} \right.
$$

where $\omega_z$ is defined as:

$$
\omega_z=\frac{\omega z}{2 \pi U}
$$


Set the parameters $A=10$, $I=0.10$, $z=30$ m, and $U=30$ m/s.

Set the target simulation length $T$ to 30 min. The time step is set to $\Delta t=0.05$ s. 

**a)** What is the neccessary frequency spacing $\Delta \omega$?

**b)** Simulate a time series using the cosine-sum method. Calculate the standard deviation of the time series, and compare it to the theoretical variance of this stochastic process.

**c)** Imagine that you now extend the time vector to last 1 hour *without* adjusting $\Delta \omega$. Explain the consequence of this choice and why this is an incorrect way to simulate longer time series.

Hint: try it and plot the result

**d)** Simulate a time series using the  IFFT method. Compare the computational time of your code to the cosine-sum method.

### Problem 3: Impulse response method and frequency response method

Consider a single degree of freedom system defined by the equation of motion: 

$$ m \ddot{y} + c \dot{y} + k y = x(t)  $$

**a)** Use the equation above to derive the impulse response method (Duhamel's integral):

$$y(t) = \int_{-\infty}^{\infty} h(t-\tau) x(\tau) d\tau $$

**b)** The stochastic load $x(t)$ has the following one-sided spectrum:

$$
S_x(\omega) 
= \left\{ \begin{array} a 10 & \quad 
0 < \omega \leq 50
\\ 0 & \quad \text{elsewhere} \end{array} \right.
$$

What do we call this type of spectrum? Use the spectrum to simulate a time series with a length of 10 minutes and a time step of 0.05 s.

**c)** Using the generated load history from question b), solve the dynamic response by the impulse response method.

**d)** Derive the frequency response method. Using the generated load history from question b) again, calculate the dynamic response using the frequency response method using the fft family of functions.
