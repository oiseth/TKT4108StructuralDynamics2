### Problem 2: Distributed loads on a long-span bridge

In this task, the objective is to obtain the response spectrum due to distributed wind loads. The considered structure is a bridge with a long span. The dynamic response considers only loads from the turbulence component $u(t)$, and the system is truncated to only include the first lateral mode $\phi(x)$.


<figure>
<img src="\Bridge.svg" align = "center" style="width:500px" >
   <figcaption align="center">  </figcaption>
</figure>

To illustrate the wind loading, we often consider long-span bridges as a continuous "beam structure" (line-like structure) subjected to a wield field consisting of a mean wind speed and turbulence:

<figure>
<img src="\Wind_field_bridge.svg" align = "center" style="width:400px" >
   <figcaption align="center">  </figcaption>
</figure>


The distributed wind load [N/m] in the linearized form can be expressed as follows:

$$
q(x,t)=\frac{1}{2} \rho D C_D V^2 + \rho D C_D V u(x,t) - \rho D C_D V \dot{r}(x,t)
$$

Set $\rho=1.2$ m, $B=3.3$ m, $C_D=0.6$, $V=30$ m/s. The bridge is $L=1300$ m long. 

The equation of motion for the single mode can be written as:

$$
\ddot{z}(t) + 2 \xi \omega_n \dot{z}(t) +  \omega_n^2 z(t) = \frac{1}{\tilde{M} } \int_0^L \phi(x) q(x,t) \: \text{d} x
$$

where the physical response is given by:
$$
r(x,t)  = \phi(x)z(t)
$$

The mode has frequency $f=0.05$ Hz and damping $\xi=0.3$%. The mode shape is $\phi(x)=\sin \left( \dfrac{x \pi}{L} \right)$ and the modal mass is $\tilde{M}=10^7$ kg.

**a)**

Neglect the static load. The modal equation of motion can be re-written as:

$$
\ddot{z}(t) + (2 \xi \omega_n-\tilde{c}_{ae}) \dot{z}(t)+  \omega_n^2 z(t) = \tilde{q}(t)
$$

Insert $q(x,t)$ into the integral and show that... 

$\quad$ *i*)  the dynamic modal load is given as follows:

$$
\tilde{q}(t) = \frac{1}{\tilde{M} } \ \rho D C_D V  \int_0^L {\phi(x) \: u(x,t) \: \text{d}x}
$$

$\quad$ *ii*) the modal aerodynamic damping is given as follows:

$$
\tilde{c}_{ae} = -\frac{1}{\tilde{M} } \rho D C_D V  \int_0^L {\phi(x)^2 \: \text{d}x}
$$


**b)** 

The auto correlation function for the modal dynamic load is given by:
$$R_{ \tilde{q} }(t_1,t_2) = E \big[ \tilde{q}(t_1) \: \tilde{q}(t_2) \big]
$$

By assuming that the wind field is stationary and homogeneous, show that the auto correlation function can be written as:
$$
R_{ \tilde{q} }(\tau) = \frac{1}{\tilde{M}^2} (  \rho D C_D V )^2 \int_0^L \int_0^L \phi(x_1) R_u(\Delta x, \tau) \phi(x_2) \: \text{d}x_1 \text{d}x_2
$$
where  $\Delta x=|x_2-x_1|$ and $R_u(\Delta x, \tau)$ is the auto correlation of the turbulence.

**c)**

The spectral density of $\tilde{q}$ is the Fourier transform of $R_{ \tilde{q} }(\tau) $: 
$$
S_{ \tilde{q} }(\omega) = \frac{1}{2 \pi}
\int_{-\infty}^{\infty} {R_{ \tilde{q} }(\tau) e^{-i \omega \tau} \: \text{d} \tau}
$$

Show that $S_{ \tilde{q} }(\omega)$ relates to  $S_{u_1 u_2}(\Delta x, \omega) $ through the following double integral:

$$
S_{ \tilde{q} }(\omega) =  \frac{1}{\tilde{M}^2} (  \rho D C_D V )^2 \int_0^L \int_0^L \phi(x_1) S_{u_1 u_2}(\Delta x, \omega)  \phi(x_2) \: \text{d}x_1 \text{d}x_2
$$

Here, $S_{u_1 u_2}(\Delta x, \omega) $ is the cross-spectral density of the vertical wind turbulence $u_1(t)$ and $u_2(t)$ at two points $x_1$ and $x_2$ along the bridge span.

**d)**

The physical response spectrum at a given position $x=x_r$ can be obtained by:

$$
S_{r_y}(\omega) \big|_{x=x_r}  = \phi(x_r)^2 S_{z}(\omega)  
= \phi(x_r)^2  |\tilde{H}(\omega)|^2 S_{ \tilde{q} }(\omega) 
$$

According to *N400 Design of bridges*, the following *single-point auto spectrum* for $u(t)$ can be used: 

$$\frac{S_u(\omega)\omega}{\sigma_u^2} = \dfrac{ \dfrac{A_u}{2\pi} \hat{\omega}_u}{\left(1+ 1.5 \dfrac{A_u}{2\pi} \hat{\omega}_u  \right)^{\dfrac{5}{3}}}
\quad , \quad
\hat{\omega}_u = \frac{\omega \: {}^xL_u}{V}$$


The cross-spectral density of the vertical wind turbulence in two given points $x_1$ and $x_2$ is given by:

$$S_{u_1 u_2}(\Delta x, \omega) = \sqrt{S_{u_1}(\omega) S_{u_2}(\omega)} \exp \left( -C_u \frac{|\Delta x| \omega}{2 \pi V} \right)
$$

Set $I_u=0.1$, $A_u=6.8$, ${}^xL_u=200$ m, and $C_u=10$.


Use the equations to numerically calculate the response spectrum in the mid span, $S_{r_y}(\omega) \big|_{x=L/2}$. Plot this spectral density for $\omega \in (0,3]$ rad/s. 

Find the corresponding standard deviation $\sigma_{r_y}\big|_{x=L/2}$ by numerical integration of the spectrum.
